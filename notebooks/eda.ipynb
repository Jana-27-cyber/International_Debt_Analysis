```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

conn = sqlite3.connect("international_debt.db")

query = """
SELECT year,
SUM(debt_value) AS total_debt
FROM debt_data
GROUP BY year
ORDER BY year
"""

df = pd.read_sql(query, conn)

plt.figure(figsize=(12,5))
plt.plot(df["year"], df["total_debt"])
plt.title("Global Debt Trend by Year")
plt.xlabel("Year")
plt.ylabel("Total Debt")
plt.grid(True)

plt.show()
```
